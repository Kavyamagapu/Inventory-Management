```python
import json
import time
fd = open("product.json",'r')
r = fd.read()
fd.close()
records = json.loads(r)
```

```python
print ("----------------------")
print("1 - View Items")
print("2 - Buy Items")
print("3 - Add new Items")
print("4 - Exit")
print ("----------------------")



def buy(ui_product,ui_quantity):              # for printing bill and updating products
    
    print("___________________________")
    print(time.strftime("           TIME : %H:%M:%S",time.localtime()))     #for printing current time

    print("___________________________")
    print("***************************")
    print("product:",records[ui_product]['product'])
    print("price:", records [ui_product]['price'])


    print("***************************")
    print("Billing Amount :",records[ui_product]['price'] *ui_quantity) 
    print("***************************")
    print("THANKS FOR VISITING!")
    print("___________________________")
    records [ui_product]['quantity']=records[ui_product]['quantity']-ui_quantity

    js = json.dumps(records)
    fd = open("product.json",'w')
    fd.write(js)
    fd.close()

ch = int(input("Enter your choice:"))

    
if(ch==1):
    #For viewing Products
    for i in records:
        print(i,records [i])

elif(ch==2):   #For buying Products
    
    ui_product = str(input("Enter the product ID:"))
    ui_quantity = int(input("Enter the quantity:"))


    if ui_quantity < records [ui_product]['quantity'] :
        buy(ui_product,ui_quantity)
    else:
        print("Sorry, We are not having that much of quantity.")
        print("We're only having",records [ui_product]['quantity'],"quantity.")
        print("Would you like to buy..?(Y/N) : ")
        c = str(input("Would you like to buy..?(Y/N) : "))
        if(c == "y" or "Y"):
            ui_quantity = int(input("Enter the quantity:"))
            buy(ui_product,ui_quantity)
        else:
            print("***************************")
            print("THANKS FOR VISITING!")
elif(ch==3):                                                           # Adding new products to inventory
    prod_id = str(input("Enter product ID:"))
    product = str(input("Enter name:"))
    price = int(input ("Enter price:"))
    quantity = int(input("Enter quantity:"))
    records[prod_id] = {'product': product,'price':price,'quantity': quantity}
    
    js = json.dumps(records)
    fd = open("product.json",'w')
    fd.write(js)
    fd.close()
    
    
else:
    print("***************************")
    print("THANKS FOR VISITING!")
    print("___________________________")    

```
